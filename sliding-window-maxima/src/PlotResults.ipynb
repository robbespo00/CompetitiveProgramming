```python
import sys
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import pandas as pd
```

```python
result_path = "results.csv"
results_df = pd.read_csv(result_path)
```

```python
sns.set_style('darkgrid')
sns.set_context('notebook')
results_df = results_df[results_df['Method'] != 'BruteForce']
results_df = results_df[results_df['Method'] != 'BruteForceIdiomatic']
for curr_n in pd.unique(results_df['n']):
    plt.title(f"n = {curr_n}")
    curr_df = results_df[results_df.n == curr_n]
    ax = sns.barplot(x="k", y="elapsed", hue="Method", data=curr_df)
    ax.legend(bbox_to_anchor=(1.0, 1.0))
    plt.show()
```
